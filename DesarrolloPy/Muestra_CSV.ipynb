```python
import pandas as pd
import os
```

```python
def dfFix(df,col1,col2):
    result = df
    x = result.columns.get_loc(col1)
    result.drop(result.columns[:x],axis = 1, inplace = True)
    y = result.columns.get_loc(col2)
    result.drop(result.columns[y+1:],axis = 1, inplace = True)
    return result
```

```python
data_Bibliography= pd.read_csv("C:/Users/guill/Documents/Universidad/PlataformaRefugiados/NAUTIA/DesarrolloPy/DataSetOriginales/Bibliography_111019.csv",sep = ";")
data_Bibliography = dfFix(data_Bibliography,"GENERAL INFORMATION - COUNTRY LEVEL","Unnamed: 3")
data_Bibliography.columns = ['GeneralInfo', 'CommunityCountry', 'RefugeeCountry']
data_Bibliography = data_Bibliography.transpose()
```
